# Matplotlib and Seaborn

Matplotlib and Seaborn are python libraries that are used for data visualization. They have inbuilt modules for plotting different graphs. While Matplotlib is used to embed graphs into applications, Seaborn is primarily used for statistical graphs.

But when should we use either of the two? Let's understand this with the help of a comparative analysis. Below are the comparison between Python's two well- known visualization packages Matplotlib and Seaborn.

# Matplotlib

- It is used for basic graph plotting like line charts, bar graphs, etc.
- It mainly works with datasets and arrays.
- Matplotlib acts productively with data arrays and frames. It regards the aces and figures as objects
- Matplotlib is more customizable and pairs well with Pandas and Numpy for Exploratory Data Analysis.

# Seaborn

- It is mainly used for statistics visualization and can perform complex visualizations with fewer commands.
- It works with entire datasets.
- Seaborn is considerably more organized and functional than Matplotlib and treats the entire dataset as a solitary unit.
- Seaborn has more inbuilt themes and is mainly used for statistical analysis.

# Line Chart

A Line chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line or curve.

Let's consider the apple yield (tons per hectare) in Kanto. Let's plot a line graph using this data and see how the yield of apples changes over time. We are going to both plot by using Matplotlib and Seaborn as follows:

# Using Matplotlib

We are using random data points to represent the yield of apples.

```python
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95] # ton per hectare
```

After we define the data points of apple yields, we are going to plot its Line Chart using Matplotlib.

```python
plt.plot(yield_apples)
```

To better understand the graph and its purpose, we can add the x-axis values too

```python
years = [2017, 2018, 2019, 2020, 2021, 2022]
yield_apples = [0.89, 0.91, 0.92, 0.93, 0.94, 0.95]
```

Since we have the 'years' variable as the x-axis, we need to type it before the yield_apples variable below

```python
plt.plot(years, yield_apples) #plt.plot(x,y)
```

Let's add labels to the axes so that we can show what each axis represents

```python
plt.plot(years, yield_apples)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
```

To plot multiple datasets on the same graph, just use the plt.plot function once fo each dataset. Let's use this to compare the yields of apples vs oranges on the same graph.

```python
apples = [0.89, 0.91, 0.93, 0.97, 0.986, 0.943]
tangarines = [0.79, 0.61, 0.93, 0.97, 0.962, 0.931]
```

```python
plt.plot(years, apples)
plt.plot(years, tangarines)
plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')
```

We can add a legend which tells us what each line in our graph means. To understand what we are plotting, we can add a title to our graph.

```python
plt.plot(years, apples)
plt.plot(years, tangarines)

plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yield in Kanto')
plt.legend(['Apples','Tangerines'])
```

To show each data point on our grap, we can higlight them with markers using the marker argument. Many different marker shapes like a circle, cross, square, diamond, etc. are provided by Matplotlib

```python
plt.plot(years, apples, marker='o')
plt.plot(years, tangarines, marker='x')

plt.xlabel('Years')
plt.ylabel('Yield (tons per hectare)')

plt.title('Crop Yield in Kanto')
plt.legend(['Apples','Tangerines'])
```

You can use the plt.figure function to change the size of the figure

```python
plt.figure(figsize=(12,6))

plt.plot(years, tangarines, marker='o')
plt.title('Yield of Tangarines (tons per hectare)')
```

# Histograms

A Histogram is a bar representation of data that varies over a range. It plots the height of the data belonging to a range along the y-axis and the range along the x-axis. Histograms are used to plot data over a range of values. They use a bar representation to show the data belonging to each range. Let's again use the 'Iris' data which contains information about flowers to plot histograms.

The 'iris' dataset is a dataset provided by Seaborn which is completely numerical.

```python
iris = sns.load_dataset('iris')
```

```python
iris
```

# Using Matplotlib

We are going to plot a histogram from the 'Sepal Width' column.

```python
iris.sepal_width
```

```python
plt.title('Distribution of Sepal Width')
plt.hist(iris.sepal_width);
```

We can control the number or size of bins too.

```python
# Specifying the number of bins
plt.hist(iris.sepal_width, bins=10);
```

# Task 3

```python
sns.set_style('darkgrid')
plt.hist(diamonds['price'], bins=30)

plt.xlabel('Harga (Price)')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga Berlian');
```

Visualize the distribution of diamond prices in a dataset. Histogram of the 'price' column in the 'diamonds' dataset using 30 bins. means that the price range will be divided into 30 different intervals and the frequency of occurrence of data in each interval will be represented in the form of bars on the histogram.

```python
import numpy as np

# specifying the number of bins
plt.hist(iris.sepal_width, bins=np.arange(2, 5, 0.25));
```

We can create bins of unequal size too.

```python
# bins of unwqual sizes
plt.hist(iris.sepal_width, bins=[1,3,4,4.5]);
```

Similar to line charts, we can draw multiple histograms in a single chart. We can reduce each histogram's opacity so that one histogram's bars don't hide the others. Let's draw separate histograms for each species of flowers. 

```python
setosa_df = iris[iris.species == 'setosa']
versicolor_df = iris[iris.species == 'versicolor']
virginica_df = iris[iris.species == 'virginica']
```

```python
plt.hist(setosa_df.sepal_width, alpha = 0.4)
plt.hist(versicolor_df.sepal_width, alpha = 0.4);
```

# Task 1

```python
diamonds = sns.load_dataset("diamonds")
```

```python
# takes the 'price' column as data
data = diamonds['price']
```

```python
sns.set_style('darkgrid')
plt.hist(data, bins=np.arange(0, 20001, 2000))

plt.xlabel('Harga (Price)')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga Berlian dengan Batas Bin yang Ditetapkan dengan NumPy');
```

```python
sns.set_style('darkgrid')
plt.hist(data, bins=[0, 1000, 3000, 5000, 8000, 20000])

plt.xlabel('Harga (Price)')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga Berlian dengan Batas Bin yang Ditetapkan');
```

```python
# Separating data based on quality (cut)
cut_groups = diamonds.groupby('cut')
```

```python
# Create a histogram for each cut group
plt.figure(figsize=(10, 6))
for name, group in cut_groups:
    plt.hist(group['price'], alpha=0.4, label=name)

plt.xlabel('Harga (Price)')
plt.ylabel('Frekuensi')
plt.title('Histogram Harga Berlian berdasarkan Kualitas (Cut)')
plt.legend(title='Kualitas (Cut)')
```

```python
plt.title('Distribution of Sepal Width')

plt.hist([setosa_df.sepal_width, versicolor_df.sepal_width, virginica_df.sepal_width], stacked = True)
plt.legend(['Setosa', 'Versicolor', 'Virginica'])
```
